```python
import pandas as pd
import datetime
import random
import numpy as np

df_before=pd.read_csv('household_power_consumption.csv', delimiter=';', low_memory=False)
df_before=df_before[df_before['Global_active_power']!='?']
names=('Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2')
names_dict={i:float for i in names}
df_before=df_before.astype(names_dict)
new = pd.to_datetime(df_before['Date'] + " " + df_before['Time'], format='%d/%m/%Y %H:%M:%S')
df_before1=pd.concat([pd.Series(new, index=df_before.index, name='Date and time'), df_before], axis=1)
df = df_before1[['Date and time', 'Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']]
df
```

```python
ar = np.array(df_before)
ar
```

5) Обрати ті рядки, які по часу >= 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини. 

```python
%%time
df_t=df[(df["Date and time"].dt.hour >= 18)&(df["Date and time"].dt.hour < 24)] 
df_act=df_t[(df_t['Global_active_power']+df_t['Global_reactive_power'])>=6.000]
df_l=df_act[(df_act['Sub_metering_2']>df_act['Sub_metering_1'])&(df_act['Sub_metering_2']>df_act['Sub_metering_3'])]
df_l
```

```python
%time df_act['Sub_metering_1'].mean(), df_act['Sub_metering_2'].mean(), df_act['Sub_metering_3'].mean()
```
